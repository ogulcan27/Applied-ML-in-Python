### Simple Regression (Basit Regresyon) problemi için veri seti oluşturma

```python
#basit regresyon problemini oluşturacak fonksiyonu import edip problem için veri seti oluşturuyor
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
plt.figure()
plt.title("Simple Regression Problem with One Input Variable")
X_R1, y_R1 = make_regression(n_samples=100, n_features=1, 
                            n_informative=1, bias=150.0,
                            noise=30, random_state=0)
plt.scatter(X_R1, y_R1, marker = "o", s=50)
plt.show()
```

+ `make_regression` fonksiyonu kullanılarak regresyon probleminin veri kümesi oluşturulur. X_R1 input, y_R1 output verileridir.
+ n_samples = 100: Oluşurulacak örnek sayısı
+ n_features = 1: Özellik sayısı (sütun sayısı)
+ n_informative = 1: Bilgilendirici özellik sayısı
+ bias = 150.0: Regresyon doğrusunun y-kesimi
+ noise = 30: Veriye eklenen gürültü miktarı
+ random_state = 0: Rastgele sayı üreteci için seed değeri

**`plt.scatter(X_R1, y_R1, marker = "o", s=50):`**

+ X_R1: x eksenindeki değerler
+ y_R1: y eksenindeki değerler
+ marker = "o": Noktaların şekli

Bu veri setini, `load_crime_dataset()` isimli bir fonksiyon ile temizleyelim ve X_crime = input, y_crime = output değerleri olarak döndürecek şekilde ayaralayalım.

```python
def load_crime_dataset():
    # Communities and Crime dataset for regression
    # https://archive.ics.uci.edu/ml/datasets/Communities+and+Crime+Unnormalized

    crime = pd.read_table('CommViolPredUnnormalizedData.txt', sep=',', na_values='?')
    # remove features with poor coverage or lower relevance, and keep ViolentCrimesPerPop target column
    columns_to_keep = [5, 6] + list(range(11,26)) + list(range(32, 103)) + [145]  
    crime = crime.iloc[:,columns_to_keep].dropna()

    X_crime = crime.iloc[:,range(0,88)]
    y_crime = crime['ViolentCrimesPerPop']

    return (X_crime, y_crime)

(X_crime, y_crime) = load_crime_dataset()
```

**`crime = pd.read_table('CommViolPredUnnormalizedData.txt', sep=',', na_values='?'):`**
 
+ Veri setini `crime` adlı değişkene atar

**`columns_to_keep = [5, 6] + list(range(11,26)) + list(range(32, 103)) + [145]:`**

+ Veri setinde kalması istenilen sütunları `columns_to_keep` değişkenine atar

**`crime = crime.iloc[:,columns_to_keep].dropna():`**

+ `crime` adlı veri setini, bütün satırları içeren ve `columns_to_keep` değişkenine atanan sütunların olduğu, ayrıca nan sütunların bırakıldığı bir veri setine dönüştürür

**`X_crime = crime.iloc[:, range(0, 88)]:`**

+ Veri setindeki ilk 88 sütunu X_crime adlı değişkene atar. Bu, input veri setini oluşturur

**`y_crime = crime['ViolentCrimesPerPop']:`**

+ "ViolentCrimesPerPop" adlı sütunu y_crime değişkenine atar. Bu, output veri setini oluşturur

**`return (X_crime, y_crime):`**

+ Oluşturulan input ve output veri setlerini bir tuple olarak döndürür

## k-Nearest Neighbors Algorithm

### Simple Binary Classification (Basit İkili Sınıflandırma) problemi için oluşturulan veri setine KNN algoritmasının uygulanması ve Overfitting ile Underfitting incelemesinin yapılması

```python
#K = 1 iken oluşturulan KNN algoritması ve doğruluğu
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test = train_test_split(X_C2, y_C2, random_state = 0) #Train ve Test veri setlerinin bölünmesi
knn = KNeighborsClassifier(n_neighbors = 1) #KNN algoritmasının oluşturulması (K = 1)
knn.fit(X_train, y_train) #KNN algoritmasının eğitilmesi
print("Test setindeki doğruluk:", knn.score(X_test, y_test)) #KNN algoritmasının doğruluğunun ölçülmesi (Test score)
print("Train setindeki doğruluk:", knn.score(X_train, y_train)) #KNN algoritmasının doğruluğunun ölçülmesi (Train score)
```

+ Görüleceği üzere K = 1 iken algoritma, Train setini çok iyi öğreniyor ve Train setindeki doğruluk, Test setindeki doğruluğa nazaran daha yüksek çıkıyor. Böylece `overfitting (aşırı uyum)` durumuna yol açıyor.

**`Algoritmanın Grafiği:`**



```python
#K = 5 iken oluşturulan KNN algoritması ve doğruluğu
X_train, X_test, y_train, y_test = train_test_split(X_C2, y_C2, random_state = 0) #Train ve Test veri setlerinin bölünmesi
knn = KNeighborsClassifier(n_neighbors = 5) #KNN algoritmasının oluşturulması (K = 5)
knn.fit(X_train, y_train) #KNN algoritmasının eğitilmesi
print("Test setindeki doğruluk:", knn.score(X_test, y_test)) #KNN algoritmasının doğruluğunun ölçülmesi (Test score)
print("Train setindeki doğruluk:", knn.score(X_train, y_train)) #KNN algoritmasının doğruluğunun ölçülmesi (Train score)
```

+ Görüleceği üzere K = 5 iken algoritma, Train setini iyi öğreniyor ancak Test setiyle doğruluk oranları K = 1'e göre daha yakın çıkıyor. Bu da modelin yeni verilere daha genelleme yapabildiğini gösteriyor.

Bu da `overfitting (aşırı uyum)` sorununa yol açıyor

Ancak hala `overfitting (aşırı uyum)` gözlenebilir.

## Linear Regression

### Simple Regression (Basit Regresyon) problemi için oluşturulan veri setine Linear Regression algoritmasının uygulanması ve Overfitting ile Underfitting incelemesinin yapılması

Simple Linear Regression (Basit Lineer Regresyon) algoritması, 1 bağımlı değişken ile 1 bağımsız değişken arasındaki ilişkiyi ifade etmek için kullanılan istatistiksel bir modelleme yöntemidir. Bu ilişki genellikle doğrusal bir denklem ile ifade edilir ve Lineer Regresyon, veri setindeki örneklerin dağılımına en iyi uyan doğruyu bulmaya çalışır. Basit Lineer Regresyon denklemi: 

$y = w_1.x_1 + b$ olarak ifade edilir.

+ $y$: Bağımlı değişken (tahmin edilen değer)
+ $w_1$: Eğim (regresyon katsayısı)
+ $x_1$: Bağımsız değişken (değişken veriler)
+ $b$: Doğrunun y eksenini kestiği nokta (regresyon sabiti, kesim noktası veya düzeltme terimi)

Bir makine öğrenmesi modelinin ne kadar iyi veya ne kadar kötü performans gösterdiğini ölçmek için `Cost Function (Maliyet Fonksiyonu)` adı verilen bir fonksiyon kullanılır. Bu fonksiyon çeşitli makine öğrenmesi modellerinde farklılık gösterebilir. Lineer regresyon için, `MSE (Mean Squared Error = En Küçük Kareler Ortalaması)` adında bir maliyet fonksiyonu kullanılır. Bu maliyet fonksiyonu, modelin tahmin ettiği değerler ile gerçek değerler arasındaki farkların karelerinin ortalamasıdır. Yani MSE fonksiyonu, modelin gerçek değerlere ne kadar uygun olduğunu gösterir. Daha düşük MSE değerleri, modelin daha iyi performans gösterdiğini gösterir. Formülü: 

$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$ şeklindedir.

+ $J({\theta})$: Maliyet fonksiyonunu temsil eder
+ $m$: Veri noktalarının sayısını temsil eder
+ $h_\theta(x^{(i)})$: Modelin i'nci örnekteki tahminini temsil eder
+ $y^{(i)}$: Gerçek i'nci örnek değerini temsil eder

Yukarıda bahsedilen maliyet fonksiyonlarını minimize etmek, modelin içindeki değerleri optimal değerlere yaklaştırmak, modelin doğruluğunu arttırır. İşte bunun için `Gradient Descent (Dereceli Azalma)` isimli bir algoritma kullanılır. Bu algoritma, bir modelin parametrelerini minimize etmek için iterarif (yinelemeli) bir yapı kullanır. Formülü: 

$\theta_j := \theta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) \cdot x_j^{(i)}$

+ $\theta_j$: J'nci regresyon katsayısı
+ $a$: Öğrenme katsayısı
+ $m$: Veri noktası sayısı
+ $h_\theta(x^{(i)})$: Modelin i'nci örnekteki tahminini temsil eder
+ $y^{(i)}$: Gerçek i'nci örnek değerini temsil eder
+ $x_j^{(i)}$: i'nci örneğin j'nci özelliğini temsil eder

Gradient Descent Algorithm (Dereceli Azalma Algoritması) dışında Lineer Regresyon denkleminin katsayılarını bulmak için `Closed-Form Solution (Normal Denklemler)` adı verilen başka bir çözüm yolu daha vardır. Bu yöntem, regresyon katsayılarını analitik bir çözümle doğrudan hesaplar.

## Ridge Regression + Normalization

### Crime dataset'ine Ridge Regression modelinin uygulanması

Ridge Regression, Linear Regression'un bir türüdür ve özellikle `multiple linear connection (çoklu doğrusal bağlantı)` sorunlarıyla başa çıkmak için kullanılır. Regresyon katsayılarını sınırmalak ve modelin overfitting (aşırı uyum) sorununu önlemek için maliyet fonksiyonuna bir `ceza terimi` ekler. Buna da `L2 regularization = L2 regülarizasyon` adı verilir. Bu, regresyon katsayılarını sıfıra yaklaştırarak modelin genelleştirilebilirliğini arttırır.

`Multiple Linear Connection (Çoklu Doğrusal Bağlantı)` sorunu, bağımsız değişkenler arasında yüksek korelasyon olduğunda ortaya çıkan bir durumdur. Bu durum, regresyon katsayılarının tahmin edilmesini zorlaştırabilir ve tahmin güvenilirliğini azaltabilir. Bu durumu engellemek için de yukarıdaki paragrafta bahsedildiği gibi `L2 düzenleme terimi` kullanılır. Bu terim, MSE formülüne bir düzenleme olarak eklenir ve aynı zamanda buna `Regularization = Regülarizasyon` da denir. Rigde regresyonun maliyet fonksiyonunun formülü:

$J(\theta) = \text{MSE}(\theta) + \lambda \sum_{i=1}^{n} \theta_i^2$ şeklindedir.

+ $J(\theta)$: Ridge regresyonunun maliyet fonksiyonu
+ $\text{MSE}(\theta)$: Mean Squared Error
+ $\lambda$: Düzenleme parametresi
+ $\theta_i$: i'nci regresyon katsayısı
+ $n$: Özellik (sütun) sayısı
+ $\lambda \sum_{i=1}^{n} \theta_i^2$: L2 düzenleme terimi = L2 Regularization

Bu formül, MSE'yi minimize ederken aynı zamanda regresyon katsayılarını küçültmeye yönelik bir düzenleme terimi içerir ve bu terim $\lambda$ ile kontrol edilir. Büyük $\lambda$ değerleri, katsayıları daha fazla küçültür ve bu da modelin daha basit olmasını sağlar. Unutmamak gerekir ki model basitleştikçe, underfitting (yetersiz uyum) problemi ortaya çıkabilir. 

Tek özellikli lineer regresyon modellerinde bu düzenleme terimi ($\lambda \sum_{i=1}^{n} \theta_i^2$) kullanılmayabilir ancak onlarca veya yüzlerce özellik içeren regresyon problemlerinde, ridge regresyon gibi düzenlenmiş lineer regresyonun kullanılmasının doğruluk iyileştirmesi önemli olabilir.

### Crime Dataset'ine uygulanan Ridge Regresyon modelinden elde edilen sonuçları iyileştirmek için Normalizasyon uygulanması

Normalizasyon, veri setindeki değerleri belirli bir aralığa veya ölçeğe getirme işlemidir. Bu işlem, veri setindeki değerlerin arasındaki ölçek farklılıklarını ortadan kaldırmayı veya azaltmayı amaçlar. Normalizasyonun kullanımındaki temel sebepler:

**`Farklı Ölçeklerdeki Değerleri Karşılaştırma:`**

Farklı özelliklerin farklı ölçeklere sahip olması durumunda, büyük değerlere sahip özellikler diğerlerini domine edebilir. Bu durumu düzeltmek için normalizasyon kullanılır, böylece tüm özellikler benzer bir ölçekte olur.

**`Model Performansını Arttırma:`**

Birçok makine öğrenimi algoritması, özelliklerin benzer ölçeklere sahip olmasını ister. Normalizasyon, modelin öğrenme sürecini ve performansını artırabilir, özellikle de gradient tabanlı optimizasyon algoritmaları kullanılıyorsa.

**`Aykırı Değerlere Karşı Dayanıklılık:`**

Normalizasyon, aykırı değerlere karşı daha dayanıklı bir model oluşturabilir, çünkü normalizasyon, genellikle değerlerin ortalamasını kullanarak ve standart sapmaya dayanarak çalıştığından, aykırı değerlerin etkisini azaltabilir.

**`Bazı Algoritmaların Performansını İyileştirme:`**

Özellikle uzaklık ölçülerine dayalı algoritmalar (örneğin, K-Nearest Neighbors algoritması), veri setindeki değerlerin ölçeklerinin benzer olmasını ister. Normalizasyon, bu tür algoritmaların performansını iyileştirebilir.

**`Optimizasyon Algoritmaları için Yardımcı:`**

Optimizasyon algoritmaları, genellikle özelliklerin benzer bir ölçekte olmasını ister. Normalizasyon, bu tür algoritmaların daha hızlı ve daha etkili çalışmasına yardımcı olabilir.

Normalizasyon yöntemleri:

**`MinMax Scaling (Min-Max Normalizasyonu):`**

Veri setindeki değerleri 0 ile 1 arasına sıkıştırır. Formülü:

$X_\text{Normalized} = \frac{X - X_\text{min}}{X_\text{max} - X_\text{min}}$ şeklindedir.

+ $X_\text{Normalized}$: Normalleştirilmiş değer
+ $X$: Orijinal değer
+ $X_\text{min}$: Özellikteki (sütundaki) minimum değer
+ $X_\text{max}$: Özellikteki (sütundaki) maximum değer

**`Z-Score Normalization (Z-Skor Normalizasyonu):`**

Veri setini standart normal dağılıma dönüştürür. Formülü:

$Z = \frac{X - \mu}{\sigma}$ şeklindedir.

+ $Z$: Z-Skor
+ $X$: Orijinal değer
+ $\mu$: Kitle ortalaması
+ $\sigma$: Standart sapma

**`Robust Scaling (Robust Normalizasyon):`**

Aykırı değerlere karşı daha dayanıklı bir normalizasyon yöntemidir. Medyan ve çeyrekler arası aralık kullanılarak yapılır. Formülü:

$X_\text{robust} = \frac{X - Medyan}{Q_3 - Q_1}$ şeklindedir.

+ $X_\text{robust}$: Normalize değer
+ $X$: Orijinal değer
+ $Medyan$: Özellikteki (sütundaki) medyan
+ $Q_1$: Özellikteki (sütundaki) 1. çeyrek
+ $Q_3$: Özellikteki (sütundaki) 3. çeyrek

**`Vector Normalization (L2 Normalizasyonu):`**

Özellik vektörlerini birim vektöre dönüştürür. Her bir özellik vektörü, bu vektörün kendi uzunluğuna bölünerek normalleştirilir. Formülü:

$X_\text{normalized} = \frac{X}{||X||_2}$ şeklindedir.

+ $X_\text{normalized}$: Normalize değer
+ $X$: Orijinal değer
+ $||X||_2$: Vektör uzunluğu

**`fit_transform() ve transform() arasındaki fark:`**

## Lasso Regression

### Crime dataset'ine Lasso Regression modelinin uygulanması

Lasso (Least Absolute Shrinkage and Selection Operator = En Küçük Mutlak Büzülme ve Seçim Operatörü) Regression, `regularization (regülarizasyon)` yöntemlerinden biridir ve çoğunlukla `çoklu regresyon problemlerinde` kullanılır. Lasso regresyonu, Ridge regresyonunda olduğu gibi büyük katsayıları sınırlamak amacıyla kullanılır ancak aynı zamanda bazı katsayıları tamamen sıfıra çekme özelliğine de sahiptir. Lasso regresyonunu, Ridge regresyonundan ayıran özellik de budur.

Lasso regresyonunda da, Ridge regresyonunda olduğu gibi maliyet fonksiyonuna eklenen bir `ceza terimi` vardır. Buna da `L1 regularization = L1 regülarizasyonu` adı verilir. Lasso regresyonunun maliyet fonksiyonu:

$J(\theta) = \text{MSE}(\theta) + \alpha \sum_{i=1}^{n} |\theta_i|$ şeklindedir.

+ $J(\theta)$: Maliyet fonksiyonu
+ $\text{MSE}(\theta)$: Mean Squared Error
+ $a$: Düzenleme parametresi
+ $\theta_i$: Regresyon katsayıları
+ $\alpha \sum_{i=1}^{n} |\theta_i|$: L1 düzenleme terimi = L1 Regularization

Lasso ve Ridge regresyonları arasındaki temel farkın, Lasso regresyonunun Ridge regresyonuna nazaran bazı katsayıları sıfıra çekme özelliğine sahip olması olduğunu yukarıdaki paragrafta yazdık. Bu fark, $\alpha \sum_{i=1}^{n} |\theta_i|$ L1 düzenleme terimindeki, $|\theta_i|$ kısmından kaynaklanır. Lasso regresyonun ceza teriminde alınan bu mutlak değer, bazı regresyon katsayılarını tamamen sıfıra çekme eğilimindedir. Bu özellik, Lasso regresyonunun `automatic feature selection (otomatik özellik seçimi)` yeteneği olarak da bilinir.

İşte bu yüzden Lasso regresyonu özellik seçimi yapmanın önemli olduğu durumlarda kullanışlıdır. Özellikle, veri setindeki özellik sayısı fazlaysa ve bunların sadece bir kısmının gerçekten tahmin için önemli olduğu düşünülüyorsa, Lasso'nun sıfıra yakın katsayıları tamamen sıfıra çekme özelliği sayesinde model daha düzenli ve etkili hale getirilebilir.

## Logistic Regression

### Simple Binary Classification (Basit İkili Sınıflandırma) problemi için oluşturulan veri setine Logistic Regression (Lojistik Regresyon) modelinin uygulanması

Lojistik Regresyon, Lineer Regresyona `Sigmoid Fonksiyonun` uygulanması sonucu elde edilen ve sınıflandırma problemleri için kullanılan bir Regresyon modelidir.  Sigmoid fonksiyonu, Lineer Regresyon modelinin sonucunda elde edilen sayısal değerleri, kendi formülü sayesinde kategorik değerlere dönüştürür. Bunu, Lineer Regresyondan elde edilen değere uyguladığı işlemler sonucu, değeri 0 ile 1 arasında bir olasılık değerine dönüştürüp eşik değerinin (genellikle 0.5 olarak kabul edilir) altında olan değeri 0, eşik değerinin üstünde olan değeri 1 olarak etiketleyerek yapar. Lojistik regresyonun fonksiyonu: 

$\frac{1}{1 + exp[-z]}$ şeklindedir.

$z = \hat{b} + \hat{w_1}*x_1 + ... + \hat{w_n}*x_n$'dir. Yani Lineer Regresyon modelidir. Özetle Lojistik Regresyon: 



```python
#Train ve Test setleri ayırıyoruz
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_C2, y_C2, random_state = 0)
```

```python
#Lojistik Regresyon modelini oluşturuyoruz
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C = 1, penalty = "l2")

#Oluşturduğumuz Modeli Eğitiyoruz
clf.fit(X_train, y_train)
```

+ `C` parametresi, Lojistik Regresyon modelinin maliyet fonksiyonunda uygulanan Regularizasyonu kontrol etmeye yarar. Bu parametre, modelin karmaşıklığını kontrol eder ve overfitting durumunu önlemeye yardımcı olur. Lojistik Regresyonun maliyet fonksiyonu:

$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(h_{\theta}(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_{\theta}(x^{(i)}))]$ şeklindedir.

+ $J(\theta)$: Maliyet fonksiyonu
+ $m$: Eğitim örneklerinin sayısı
+ $y^{(i)}$: i'nci eğitim örneğinin gerçek sınıf etiketi (0 veya 1)
+ $h_{\theta}(x^{(i)})$: i'nci eğitim örneği için Lojistik regresyonun tahmin ettiği olasılık değeri

Dikkat edersek $y^{(i)} = 0$ iken formülün $(1 - y^{(i)}) \log(1 - h_{\theta}(x^{(i)}))]$ kısmı hesaplanıyor ve $y^{(i)} = 1$ iken formülün $y^{(i)} \log(h_{\theta}(x^{(i)}))$ kısmı hesaplanıyor.


+ `penalty` parametresi, uygulanacak olan Regularizasyonu belirler. Default olarak ``l2 (L2 Regülarizasyondur)`` ancak ``l1`` veya ``None`` olarak değiştirilebilir. Ridge ve Lasso regresyondaki gibi Regularizasyon uygular.

```python
#Eğittiğimiz modelin Train ve Test setlerindeki doğruluğuna bakıyoruz
print("Train setindeki doğruluk:", clf.score(X_train, y_train))
print("Test setindeki doğruluk:", clf.score(X_test, y_test))
```

## Support Vector Machines

### Complex Binary Classification (Karmaşık İkili Sınıflandırma) problemi için oluşturulan veri setine Linear Support Vector Machines (Lineer Destek Vektör Makineleri) algoritmasının uygulanması ve $C$ parametresine göre Overfitting ile Underfitting incelemesinin yapılması

Support Vector Machines (Destek Vektör Makineleri), hem sınıflandırma hem de regresyon problemlerine uygulanabilen güçlü bir ML algoritmasıdır. Özellikle sınıflandırma problemlerinde yaygın olarak kullanılır. Sınıflandırma problemleri için **lineer** ve **non-lineer** SVM algoritmaları vardır.

Linear Support Vector Machines (Doğrusal Destek Vektör Makineleri), sınıflandırma problemleri için kullanılabilen **lineer** bir modeldir. Veri noktalarını sınıflandırmak için bir `hiperdüzlem` belirler ve bu `hiperdüzlemi` optimize etmeye çalışır. `Hiperdüzlem`, bir doğru (bu doğruya karar sınırı adı verilir) ile sınıfları birbirinden ayırır ve iki sınıfın noktalarının bu doğruya maksimum uzaklıkta olmasını sağlamaya çalışır. Bu maksimum sınır ise `margin (marj)` doğruları ile belirlenir ve bu mantık, aynı zamanda `Large Margin Classification` olarak da adlandırılır. 

Veri noktaları $(x_1, y_1), (x_2, y_2), ..., (x_n, y_n)$ olmak üzere, Lineer SVM formülü;

$f(x) = sign(\vec{w}* \vec{x} + b)$ yani $f(x) = \sum{w[i]*x[i]} + b$ şeklindedir. 

+ $sign()$: $\vec{w}* \vec{x} + b < 0$ ise veri noktasının etiketini `-1` olarak ayarlayan ya da $\vec{w}* \vec{x} + b \geq 0$ ise veri noktasının etiketini `1` olarak ayarlayan fonksiyondur.
+ $\vec{w}$: Her bir özelliğin (sütunun) katsayısıdır

+ Sınırlara değen veri noktaları, `Support Vector (Destek Vektörü)` olarak adlandırılır ve `margin (marj)` sınırının belirlenmesine yardımcı olur.

### Complex Binary Classification (Karmaşık İkili Sınıflandırma) problemi için oluşturulan veri seti üzerinde Kernelized Support Vector Machines (Kernelize Destek Vektör Makineleri) algoritmalarının uygulanması 

Kernelized SVM, doğrusal olmayan veri setlerini işlemek için kullanılırlar ve **non-lineer** modellerdir. Veri noktalarının dağılımı doğrusal olarak ayrılabilir olmadığında Kernelized SVM, girdi (input) özelliklerini yüksek boyutlu uzayda yeniden haritalayan ``Kernel`` fonksiyonlarını kullanır ve verilerin doğrusal olarak ayrılabilir hale gelmesini sağlar. Burada aslında veriler fiziksel olarak bir değişime uğramazlar, sadece düşük boyutlu uzayda ayrılamaz olarak görünürlerken, daha yüksek boyutlu bir uzaya yansıtılıp orada ayrılabilir olarak görünmleri sağlanır.

En yaygın kullanılan kernel fonksiyonları arasında `Polynomial`, ``RBF (radial basis function)``, ``Sigmoid`` vb. bulunur. Bu kernel fonksiyonları, veriye göre seçilebilir ve veri setinin yapısına uygun olarak tercih edilebilir.

Kernelize edilmiş SVM'ler, genellikle doğrusal olmayan veri setlerinde, özellikle de yüksek boyutlu veri setlerinde ve karmaşık veri yapılarında iyi performans gösterirler. Ancak, parametre ayarlaması ve hesaplama maliyeti gibi bazı zorluklar da olabilir.

## Cross Validation 

Cross Validation (Çapraz Doğrulama), bir modelin performansını değerlendirmek ve test etmek için aynı veri setini birden fazla kez kullanma işlemidir. Temel amacı, modelin öğrenme ve genelleme yapma yeteneğini ölçmektir. Veri setini rastgele parçalara ayırarak her parçayı sırayla test ve eğitim verisi olarak kullanır. Böylece model, veri setindeki her veriyi hem Train hem de Test için görmüş olur. Arka planda, oluşturulan her Train ve Test seti için ayrı bir eğitimli model oluşturulur ve her model için doğruluk hesaplanır. Ardından oluşturulan modellerden elde edilen doğruluk değerlendirmelerinin ortalaması alınır. 

Dolayısıyla sadece Train ve Test setlerindeki doğruluk oranına bakarak bir modelin öğrenme ve genelleme yeteneği hakkında yorum yapmak, Cross Validation'a göre daha az güvenilirdir.

## Decision Trees

Decision Trees (Karar Ağaçları), veri setindeki özelliklere dayalı olarak veri noktalarını sınıflandırmak veya değerlerini tahmin etmek için sınıflandırma ve regresyon problemlerinde kullanılırlar. Veri setindeki özelliklerin değerlerine göre ardışık kararlar alarak bir ağaç yapısı oluştururlar.

Karar ağaçlarının teorik mantığı:

+ **Root Node (Kök düğüm):** Karar ağacının başlangıç noktası. Tüm veri seti bu düğüme gelir.
+ **Feature Selecting and Splitting (Özellik Seçimi ve Bölme):** Her düğümde, ağaç, veri setini belli bir özelliğin değerine göre böler. Özellikle, veri setindeki özelliklerin her biri için en iyi bölme noktası seçilmeye çalışılır. Bölme noktası, veri noktalarını mümkün olduğunca homojen alt gruplara ayırmak için belirlenir. Bu bölme işlemi, veri setini daha küçük ve daha homojen alt gruplara bölen bir ağaç yapısı oluşturur.
+ **Child Nodes (Alt Dallar):** Her bölme, yeni bir düğüm oluşturur ve bu düğümler alt dallar olarak adlandırılır. Karar ağacının her bir düğümü, bir özellik değerine göre alt dallara bölünmüş veri alt kümelerini temsil eder.
+ **Branching (Dallanma):** Ağaç, her özellik için bir dallanma yapar ve bu dallanmalar, veri setinin farklı özellik değerlerine göre ayrılmasını sağlar.
+ **Leaf Nodes (Yaprak Düğümler):** Ağacın sonunda, homojen bir alt grup oluşturulduğunda veya belirli bir duruma ulaşıldığında, yaprak düğümler oluşur. Yaprak düğümler, sınıflandırma problemlerinde belirli bir sınıfa veya regresyon problemlerinde tahmin edilen bir değere karşılık gelir.

**`c_features = len(feature_names):`**

+ Özellik sayısını (sütun sayısı) `c_features` isimli değişkene atar.

**`plt.barh(range(c_features), clf.feature_importances_):`**

## One-Hot Encoding

One-Hot Encoding, kategorik verileri, sayısal formata dönüştürmek için kullanılan bir tekniktir. Bu teknik, makine öğrenimi modellerinin kategorik verilerle çalışmasına olanak tanır. Her kategorik değişken değeri için yeni bir sütun oluşturur ve bu sütunlardan biri 1'i, diğerleri 0'ı temsil eder.  Dolayısıyla, her kategorik değer, kendi benzersiz sütununda temsil edilir.

Örneğin, "renk" adında bir kategorik değişkenimiz varsa ve bu değişkenin değerleri "kırmızı", "mavi" ve "yeşil" ise, One-Hot Encoding şu şekilde uygulanır:


One-Hot Encoding'in amacı, makine öğrenimi modellerinin, kategorik değişkenlerin sıralamasını veya büyüklüğünü yanlış yorumlamasını engellemektir. Ayrıca, kategorik değişkenlerin sayısal formata dönüştürülmesi, modele kategorik bilgiyi daha etkin bir şekilde iletebilir., 1]

```python
import numpy as np
import pandas as pd
np.random.seed(0)
data = {"number_1": np.random.rand(15),
       "number_2": np.random.rand(15),
       "categorical": np.random.choice(["A", "B", "C", "D"], size = 15)}

df = pd.DataFrame(data)
df
```

```python
#Veri setini Train ve Test olarak bölebilmek için input (number_1 ve number_2 sütunları)
#ve output (categorical) sütununu ayırıyoruz
X = df[["number_1", "number_2"]]
y = df[["categorical"]]

print(X)
print("\n", y)
```

`categorical` sütununa Scikit-Learn ve Pandas ile One-Hot Encoding uygulayalım. 

Bu 4 sütunu 1,2,3,4 olarak kategorilendirebiliriz ancak modeli eğittiğimizde bu değerleri ağırlıklandıracağından dolayı yanlış bir sonuç verecektir. Bu yüzden One-Hot Encoding uygulayarak sınıflar arasında sayısal bir fark oluşmadan, kategorik verileri sayısal bir veriye dönüştüreceğiz.

### Scikit-Learn ile One-Hot Encoding

```python
#OneHotEncoder oluşturma ve uygulama
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
y_encoded = encoder.fit_transform(y).toarray()

#Dönüştürülmüş veriyi DataFrame'e dönüştürme
encoded_df = pd.DataFrame(y_encoded, columns=encoder.get_feature_names_out(['categorical']))

#Sayısal sütunları yeni veri setine ekleyin
encoded_df[['number_1', 'number_2']] = df[['number_1', 'number_2']]

encoded_df
```
